```python
#Created by RM
#This notebook is primery designed to test statistical and exploratory techniques in
#NLP like distributional properties, feature sparsity, and dimensionality

#Zipf's Law
# Purpose: Shows the frequency distribution of words
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
data=pd.read_csv('bbc_data.csv')
all_words = ' '.join(data['data']).split()
word_counts = Counter(all_words)
common_words = word_counts.most_common(50)

words, counts = zip(*common_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.xticks(rotation=90)
plt.title('Word Frequency (Top 50)')
plt.savefig('Word Frequency (Top 50)')
plt.show()
```

```python
#TF-IDF Sparsity:  Measures sparsity in feature space .
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(data['data'])
sparsity = (X.nnz / float(X.shape[0] * X.shape[1])) * 100
print(f"TF-IDF Matrix Sparsity: {sparsity:.2f}%")
```

```python
#Vocabulary Size: Understands the richness of the text by calculating unique word counts.
unique_words = len(set(all_words))
print(f"Vocabulary size: {unique_words}")
```
